**Cleaning the Data**
1. URLs
2. Hashtags
3. Mentions
4. Special Letters
5. Punctuations

```python
import re

def cleanResume(txt): #Cleans the input and returns only the text by removing extra letters/characters
    cleanTxt = re.sub(r'http\S+\s',' ',txt) # Remove URLs
    cleanTxt = re.sub(r'@\S+',' ',cleanTxt) # Remove @mentions 
    cleanTxt = re.sub(r'#\S+\s',' ',cleanTxt) # Remove hashtags
    cleanTxt = re.sub(r'RT|cc',' ',cleanTxt)  # Remove RT and cc
    cleanTxt = re.sub(r'[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\\]^_`{|}~"""),' ',cleanTxt)  # Remove special characters
    cleanTxt = re.sub(r'[^\x00-\x7f]',' ',cleanTxt) # Remove non-ASCII characters
    cleanTxt = re.sub(r'\s+',' ',cleanTxt) # Remove extra whitespaces
    return cleanTxt
```

```python
df['Resume']
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

**Words into Categorical Values**

This dataset is supervised because it includes labeled data, where each resume (input feature) is associated with a specific category (output label), providing clear guidance for the model during training. Additionally, it's multilevel categorical data because the "category" column contains multiple distinct levels or classes, representing various professions or fields such as Data Science, Web Development, and Java Development, among others.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category'] = le.transform(df['Category'])
```

**Vectorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer #Term frequency inverse document frequency 
tfidf = TfidfVectorizer(stop_words='english')

tfidf.fit(df['Resume'])
requiredtext = tfidf.transform(df['Resume'])

# This code uses a technique called TF-IDF to convert text data from resumes into numerical vectors. First, it learns the unique words from the resumes 
# and their importance across all resumes. Then, it transforms each resume into a set of numbers, where each number represents the importance 
# of a specific word within that resume compared to its importance across all resumes.
```

```python
df
```

**Splitting**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(requiredtext, df['Category'], test_size=0.2, random_state=42)
#In this code, 20% of the data is allocated for testing (X_test and y_test), while the remaining 80% 
#is used for training (X_train and y_train). So, the data split is 80% training and 20% testing.
```

```python
X_train.shape
```

```python
X_test.shape
```

**Training the Model and Printing the Classification Report:**
Will be using KNN(K-Nearest Neighbor classifier)

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train,y_train)
ypred = clf.predict(X_test)
print(ypred)
print(accuracy_score(y_test,ypred))
```
